Data Quality Check for Benin

Data Quality Check for Togo

```python
columns_to_modify = ['GHI', 'DNI', 'DHI']
togo[columns_to_modify] = togo[columns_to_modify].clip(lower=0)
print(togo.head())
```

Time Series Analysis for Benin

```python
benin['Timestamp'] = pd.to_datetime(benin['Timestamp'])
benin.set_index('Timestamp', inplace=True)
fig, axes = plt.subplots(nrows=4, figsize=(10, 12))
variables = ['GHI', 'DNI', 'DHI', 'Tamb']
for i, var in enumerate(variables):
    axes[i].plot(benin.index, benin[var])
    axes[i].set_ylabel(var)

plt.xlabel('Timestamp')
plt.tight_layout()
plt.show()
```

Time Series Analysis for Sierraleon

```python
sierra['Timestamp'] = pd.to_datetime(sierra['Timestamp'])
sierra.set_index('Timestamp', inplace=True)
fig, axes = plt.subplots(nrows=4, figsize=(10, 12))
variables = ['GHI', 'DNI', 'DHI', 'Tamb']
for i, var in enumerate(variables):
    axes[i].plot(sierra.index, sierra[var])
    axes[i].set_ylabel(var)

plt.xlabel('Timestamp')
plt.tight_layout()
plt.show()
```

Correlation Analysis of Benin

```python
solar_columns = ['GHI', 'DHI', 'DNI']
temperature_columns = ['TModA', 'TModB']
selected_data = benin[solar_columns + temperature_columns]

correlation_matrix = selected_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
solar_columns = ['GHI', 'DHI', 'DNI']
temperature_columns = ['TModA', 'TModB']
selected_data = sierra[solar_columns + temperature_columns]

correlation_matrix = selected_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

Histograms for Benin

```python
variables_of_interest = ['GHI', 'DNI', 'DHI', 'WS', 'TModA', 'TModB']
fig, axes = plt.subplots(nrows=len(variables_of_interest), ncols=1, figsize=(8, 6 * len(variables_of_interest)))
for i, variable in enumerate(variables_of_interest):
    ax = axes[i]
    ax.hist(benin[variable], bins=20, edgecolor='black')
    ax.set_xlabel(variable)
    ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
variable_pairs = [('GHI', 'Tamb'), ('WS', 'WSgust'), ('ModA', 'WD')]
fig, axes = plt.subplots(nrows=len(variable_pairs), ncols=1, figsize=(8, 6 * len(variable_pairs)))
for i, (var1, var2) in enumerate(variable_pairs):
    ax = axes[i]
    ax.scatter(benin[var1], benin[var2], alpha=0.5)
    ax.set_xlabel(var1)
    ax.set_ylabel(var2)

plt.tight_layout()
plt.show()
```
